```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Seasons_Stats.csv')
```

```python
df=df[df['Year']==2017.0]
```

```python
df.info()
```

```python
df.drop(columns=['blanl','blank2'], inplace=True)
```

```python
df.drop_duplicates(subset='Player', inplace=True)
```

```python
df.Tm.unique()
```

```python
df.fillna(0, inplace=True)
```

### Who is the youngest and oldest player in the NBA in 2017 for each team (Tm) ?

```python
df['group_rank_max']=df.groupby('Tm')['Age'].rank(ascending=0, method='dense')
df['group_rank_min']=df.groupby('Tm')['Age'].rank(ascending=1, method='dense')
```

```python
max_a=df[df['group_rank_max']==1.0][['Player','Age', 'Tm']].set_index('Tm').sort_index()
max_a.reset_index(inplace=True)
max_a.rename(columns={'Tm':'Team','Player':'Oldest Player'})
```

```python
min_a=df[df['group_rank_min']==1.0][['Player','Age', 'Tm']].set_index('Tm').sort_index()
min_a.reset_index(inplace=True)
min_a.rename(columns={'Tm':'Team','Player':'Youngest Player'})
```

### Which player has the most minutes played (MP) in each position (Pos)?

```python
df['most_played_max'] = df.groupby('Pos')['MP'].rank(ascending=0, method='dense')
dfMP = df[df['most_played_max']==1.0][['Player','MP', 'Pos', 'Tm']].set_index('Pos')
dfMP.reset_index(inplace=True)
dfMP.rename(columns={'MP':'Minutes Played','Tm':'Team', 'Pos':'Position'}, inplace=True)
dfMP.sort_values(by='Minutes Played')
```

### Who is the best player based on his record stats?

```python
best_player =df[['Player','Tm','TRB%','FT%', 'TOV%', 'TS%', 'STL%', 'BLK%']]
```

```python
best_player
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
best_player[['TRB%','FT%','TOV%','TS%','STL%','BLK%']] = scaler.fit_transform(best_player[['TRB%','FT%','TOV%','TS%','STL%','BLK%']])
```

```python
best_player.head()
```

```python
best_player.fillna(0, inplace=True)
best_player['score']=best_player['TRB%']*0.15+best_player['TS%']*0.3+best_player['STL%']*0.1+best_player['FT%']*0.1+best_player['TOV%']*0.25+best_player['BLK%']*0.1
```

```python
best_player.sort_values(by='score', ascending=False)
```

```python
best_team = best_player.groupby('Tm')[['score']].mean()
```

```python
best_team.reset_index(inplace=True)
```

```python
best_team.sort_values(by='score', ascending=False)
```
